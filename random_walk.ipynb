```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
n=100000
```

```python
x=np.zeros(n)
y=np.zeros(n)
```

```python
for i in range(1,n):
  xwalk = 0
  ywalk = 0
  xory = random.choice([0, 1]) # if 0 x more, if 1 y move
  if xory==1:
    ywalk = random.choice([1, 0, -1])
  else:
    xwalk = random.choice([1, 0, -1])
  x[i]=x[i-1]+xwalk
  y[i]=y[i-1]+ywalk
```

```python
plt.figure(figsize=(20,10))
plt.title("Random Walk")
plt.plot(x, y)
plt.plot(x[0],y[0],color='green',marker='o')
plt.plot(x[n-1],y[n-1],color='red',marker='o')
```
